```python
%pylab inline
import pandas
import numpy
import seaborn
```

# Load CSV file into memory

```python
data = pandas.read_csv('D:/Programming Projects/fitbit-data-analysis/health-data.csv')
```

```python
data['date'] = data['date'].map(pandas.to_datetime)
```

### Add some useful columns

```python
def get_dom(dt):
    return dt.day

data['day_of_month'] = data['date'].map(get_dom)
```

```python
def get_weekday(dt):
    return dt.weekday()

data['weekday'] = data['date'].map(get_weekday)
```

```python
def get_month(dt):
    return dt.month

data['month'] = data['date'].map(get_month)
```
